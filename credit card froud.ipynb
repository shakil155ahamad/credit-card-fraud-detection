```python
print(froud.shape)
nofroud1=no_froud1.sample(frac=1)
froud=froud.sample(frac=1)
no_froud_equal=no_froud1[no_froud1['Class']==0][:492]
print(no_froud_equal.shape)
df_equal=pd.concat([froud,no_froud_equal])
df_equal=df_equal.sample(frac=1,random_state=42)
df_equal.head(10)
```

# standradization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_time=scaler.fit_transform(df_equal['Time'].values.reshape(-1,1))
scaled_amount=scaler.fit_transform(df_equal['Amount'].values.reshape(-1,1))
df_s=df_equal.copy()
df_s=df_s.drop(['Amount','Time'],axis=1)
df_s.insert(0,'Scaled Amount',scaled_amount )
df_s.insert(1,'Scaled Time',scaled_time )
df_s.head()
```

```python
fx,ax=plt.subplots(1,2,figsize=(10,5))
sns.distplot(df_s['Scaled Time'].values,ax=ax[0],color='g')
ax[0].set_title('time')
sns.distplot(df_s['Scaled Amount'].values,ax=ax[1],color='r')
ax[1].set_title('Amount')
plt.show()
```

# splitting of data in input and target

```python
x=df_s.drop('Class',axis=1)
y=df_s['Class']
print(x.shape)
print(y.shape)
```

```python
from sklearn.decomposition import PCA,TruncatedSVD
from sklearn.manifold import TSNE
pca=PCA(n_components=2,random_state=42)
pca_components=pca.fit_transform(x)
tsne=TSNE(n_components=2)
tsne_components=tsne.fit_transform(x)
tsne_components.shape
svd=TruncatedSVD(n_components=2,random_state=42)
svd_components=svd.fit_transform(x)
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(12,5))
ax1.scatter(pca_components[:,0],pca_components[:,1],c=(y==0),label='no_froud',cmap='rainbow')
ax1.scatter(pca_components[:,0],pca_components[:,1],c=(y==1),label='froud',cmap='rainbow')
ax1.set_title('PCA_components')
ax1.grid(True)
ax1.legend()

ax2.scatter(tsne_components[:,0],tsne_components[:,1],c=(y==0),label='no_froud',cmap='rainbow')
ax2.scatter(tsne_components[:,0],tsne_components[:,1],c=(y==1),label='froud',cmap='rainbow')
ax2.set_title('tsne_components')
ax2.grid(True)
ax2.legend()

ax3.scatter(svd_components[:,0],svd_components[:,1],c=(y==0),label='no_froud',cmap='rainbow')
ax3.scatter(svd_components[:,0],svd_components[:,1],c=(y==1),label='no_froud',cmap='rainbow')
ax3.set_title('svd_components')
ax3.grid(True)
ax3.legend()

plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    'Support vector classifers':SVC(C=1,kernel='linear'),
     "KNearest": KNeighborsClassifier( n_neighbors=5),
     'randomforest':RandomForestClassifier()
}
```

```python
for key, clf in classifiers.items():
    clf.fit(x_train, y_train)
    training_score = cross_val_score(clf, x_train, y_train,cv=10)
    print("Classifiers: ", clf.__class__.__name__, "Has a training score of", round(training_score.mean(),3) * 100, "% accuracy score")
```

# selecting the best parameter

```python
from sklearn.svm import SVC
model_params={
               'svm':{
                       'model':SVC(gamma='auto'),
                        'params':{'C':list(range(2,10,1)),'kernel':['rbf','linear','poly']}
               },
                'LogisticRegression':{
                                      'model':LogisticRegression(),
                                       'params':{'penalty': ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
                },
                 'Knearestneighbour(knn)':{
                                          'model':KNeighborsClassifier(),
                                            'params':{ 'n_neighbors':list(range(2,10,1)),'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute']}
                     
                 },
                  'RandomForestClassifier':{
                                           'model':RandomForestClassifier(),
                                            'params':{'n_estimators':list(range(2,7,1))}
                  } 
                   
               
    
    
}
```

```python
from sklearn.model_selection import GridSearchCV
best_est=[]
score=[]
for model_name,key in model_params.items():
    clf=GridSearchCV(key['model'],key['params'],cv=10,return_train_score=False)
    clf.fit(x_train,y_train)
    score.append({'model':model_name,
                   'best_params':clf.best_params_,
                    'best_score':clf.best_score_
                     })
    best_est.append(clf.best_estimator_)
    
```

```python
score=pd.DataFrame(score)
score
```

# analysis of test data using confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pre_svm_cf=confusion_matrix(best_est[0].predict(x_test),y_test)

y_pre_log_cf=confusion_matrix(best_est[1].predict(x_test),y_test)

y_pre_knn_cf=confusion_matrix(best_est[2].predict(x_test),y_test)

y_pre_rf_cf=confusion_matrix(best_est[3].predict(x_test),y_test)
```

```python
fig = plt.figure(figsize = (8,8)) # width x height
ax1 = fig.add_subplot(2, 2, 1) # row, column, position
ax2 = fig.add_subplot(2,2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
sns.heatmap(y_pre_svm_cf,annot=True,cmap='BuGn',ax=ax1)
ax1.set_title('svm\n confusion matrix')
sns.heatmap(y_pre_log_cf,annot=True,cmap='BuGn',ax=ax2)
ax2.set_title('logistic\n confusion matrix')
sns.heatmap(y_pre_knn_cf,annot=True,cmap='BuGn',ax=ax3)
ax3.set_title('knn \n confusion matrix')
sns.heatmap(y_pre_rf_cf,annot=True,cmap='BuGn',ax=ax4)
ax4.set_title('random forest\n confusion matrix')
plt.show()
```
